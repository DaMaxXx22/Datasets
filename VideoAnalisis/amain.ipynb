```python
import pandas as pd
import numpy as np
from IPython.display import display as dp
import os
```

```python
base='documents'
ventas=pd.read_csv(os.path.join(base,'Ventas.csv'))
inventario=pd.read_csv(os.path.join(base,'Inventario.csv'))
productos=pd.read_csv(os.path.join(base,'Productos.csv'))
tiendas=pd.read_csv(os.path.join(base,'Tiendas.csv'))
```

```python
dp(ventas.head())
```

```python
dp(inventario.head())
```

```python
dp(tiendas.head())
```

```python
dp(productos.head())
```

```python
df_analisis=ventas.merge(inventario, on='ProductoID', how='left')
```

```python
dp(df_analisis.head())
```

```python
df_analisis=df_analisis.merge(tiendas, on='TiendaID', how='left')
```

```python
dp(df_analisis.head())
```

```python
df_analisis=df_analisis.merge(productos, on=['ProductoID','PrecioUnitario'], how='left')
```

```python
dp(df_analisis.head())
```

```python
df_analisis.describe()
```

```python
df_analisis.info()
```

```python
df_analisis['Fecha']=pd.to_datetime(df_analisis['Fecha'])
```

```python
df_analisis.info()
```

```python
df=df_analisis.copy()
```

```python
df['Mes']=df['Fecha'].dt.month
df['Mes_Nombre']=df['Fecha'].dt.month_name()
```

```python
dp(df.head())
```

```python
ingresos_por_mes=df.groupby(['Mes','Mes_Nombre','Canal'])['Ingreso'].sum().reset_index().drop(columns='Mes')
ingresos_por_mes.columns=['Mes','Canal','Ingreso']
```

```python
dp(ingresos_por_mes)
```

```python
p_vendidos_ingreso=df.groupby('NombreProducto')['Ingreso'].sum().reset_index()
top_10=p_vendidos_ingreso.nlargest(10,'Ingreso').reset_index()
```

```python
dp(top_10)
```

```python
p_vendidos_ingreso=df.groupby('NombreProducto')['Cantidad'].sum().reset_index()
top_10_cant=p_vendidos_ingreso.nlargest(10,'Cantidad').reset_index()
```

```python
dp(top_10_cant)
```

```python
df['Margen']=df['PrecioUnitario']-df['CostoUnitario']
```

```python
dp(df.head())
```

```python
mc=df.groupby('Categoría')['Margen'].mean().reset_index()
```

```python
dp(mc)
```

```python
rf=df[(df['StockActual'] < 50) & (df['Ingreso'] > 0)]
rf=rf[['NombreProducto','StockActual','Ingreso']].reset_index().drop(columns='index')
rf=rf.groupby('NombreProducto').agg({
    'StockActual':'sum',
    'Ingreso':'sum'
})
```

```python
dp(rf)
```

```python
mf=df.groupby(['NombreTienda','Ciudad','Mes','Mes_Nombre'])['Ingreso'].sum().reset_index().drop(columns='Mes')
mf.columns=['Tienda','Ciudad','Mes','Ingresos']
```

```python
dp(mf.head())
```

```python
df.to_csv('Analisis_PT.csv',index=False,encoding='utf-8-sig')
```
