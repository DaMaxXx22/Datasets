```python

def mp_me(ruta):
    #Respuestas Esperadas
    re={
    'Titulado':'si',
    'Puesto Actual':'data analyst',
    'Sueldo Esperado': '23000',
    'Ciudad':['cdmx', 'ciudad de mexico'],
    'Automovil':'si',
    'Ultimo grado de Estudio':'Licenciatura'
}
    try:
        df_mp=pd.read_excel(ruta,dtype='str')
    except Exception as e:
        df_mp=pd.read_csv(ruta,dtype='str')
    
    def Validacion():
        columnas=list(df_mp.columns)[3:]
        scores=[]
        for i in columnas:
            score=[]
            for j in df_mp[i]:
        
                if j.lower() == re[i]:
                    score.append(1)
                else:
                    score.append(0)
            scores.append(score) 

        calificaciones=[]

        for i in range(0,len(df_mp)):
            calificacion=0
            for j in scores:
                calificacion+=j[i]
            calificaciones.append(calificacion)
        
        df_mp['Score']=calificaciones
        df_ps=df_mp[df_mp['Score'] == df_mp['Score'].max()].reset_index(drop=True)
        return df_ps.to_excel('Posibles Candidatos.xlsx', index=False)
    return Validacion()
    
```
