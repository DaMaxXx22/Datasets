```python
import pandas as pd
import os
from IPython.display import display

```

```python
df1=pd.read_csv(os.path.join('..','DataFrames','archivo1.csv'))
df2=pd.read_csv(os.path.join('..','DataFrames','archivo2.csv'))
```

```python
display(df1.head())
```

```python
display(df2.head())
```

```python
outer=pd.merge(df1,df2,on='ID',how='outer',suffixes=('_izquierda','_derecha'))
```

```python
display(outer.head(6))
```

```python
outer=pd.merge(df1,df2,on='ID',how='inner',suffixes=('_izquierda','_derecha'))
```

```python
display(outer.head(6))
```

```python
outer=pd.merge(df1,df2,on='ID',how='left',suffixes=('_izquierda','_derecha'))
```

```python
display(outer.head(6))
```

```python
outer=pd.merge(df1,df2,on='ID',how='right',suffixes=('_izquierda','_derecha'))
```
