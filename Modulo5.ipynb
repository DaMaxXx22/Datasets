**Formular las Hipótesis**

**Hipótesis 1: Comparar las calificaciones promedio de los usuarios para Xbox One y PC.**

- Hipótesis Nula (H₀): Las calificaciones promedio de los usuarios para Xbox One y PC son iguales. Es decir,  𝜇Xbox One=𝜇PC

- Hipótesis Alternativa (H₁): Las calificaciones promedio de los usuarios para Xbox One y PC son diferentes. Es decir,  𝜇Xbox One≠𝜇PC

**Hipótesis 2: Comparar las calificaciones promedio de los géneros de Acción y Deportes**

- Hipótesis Nula (H₀): Las calificaciones promedio de los usuarios para los géneros de Acción y Deportes son iguales. Es decir, 𝜇Accion≠𝜇Deportes 

- Hipótesis Alternativa (H₁): Las calificaciones promedio de los usuarios para los géneros de Acción y Deportes son diferentes. Es decir, 𝜇Accion=𝜇Deportes



```python
alfa=0.05
```

```python
#import scipy.stats as stats


xru=df[df['platform']=='XOne']['user_score']
pcru=df[df['platform']=='PC']['user_score']

aru=df[df['genre']=='Action']['user_score']
dru=df[df['genre']=='Sports']['user_score']

#con la prueba T
t1,r1=stats.ttest_ind(xru,pcru,equal_var=False)
print(f'HIPOTESIS 1 (XONE VS PC): T_result: {t1} -- result: {r1}')


t2,r2=stats.ttest_ind(aru,dru,equal_var=False)
print(f'HIPOTESIS 2 (Action VS Sports): T_result: {t2} -- result: {r2}')

# Interpretación de los resultados
if r1 < alfa:
    print("Rechazamos H₀ para la Hipótesis 1: Las calificaciones promedio de los usuarios para Xbox One y PC sean iguales.")
else:
    print("No rechazamos H₀ para la Hipótesis 1: No hay evidencia suficiente para decir que las calificaciones promedio para Xbox One y PC sean iguales.")

if r2 < alfa:
    print("Rechazamos H₀ para la Hipótesis 2: Las calificaciones promedio de los usuarios para los géneros Acción y Deportes son diferentes.")
else:
    print("No rechazamos H₀ para la Hipótesis 2: No hay evidencia suficiente para decir que las calificaciones promedio para Acción y Deporte")

```
